Sales Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Downloads/Moumita/AusApparalSales4thQrt2020.csv"
df = pd.read_csv(file_path)

# Aggregate total sales by state
sales_by_state = df.groupby("State")["Sales"].sum().sort_values(ascending=False)

# Display the sales by state
print("=== Total Sales by State ===")
print(sales_by_state)

# Identify top and low-performing states
top_states = sales_by_state.head(3)
low_states = sales_by_state.tail(3)

print("\n=== Top-Performing States ===")
for state, sales in top_states.items():
    print(f"{state}: ${sales:,.2f}")

print("\n=== Low-Performing States ===")
for state, sales in low_states.items():
    print(f"{state}: ${sales:,.2f}")

# Plot the sales by state
plt.figure(figsize=(10, 6))
sales_by_state.plot(kind="bar", color="skyblue")
plt.title("Total Sales by State (Q4)")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Recommendations for low-performing states
print("\n=== Recommendations for Low-Performing States ===")
for state in low_states.index:
    print(f"\nState: {state}")
    print("- Launch localized marketing campaigns")
    print("- Offer seasonal discounts and bundle promotions")
    print("- Focus on Kids and Women categories to boost engagement")
    print("- Explore opportunities for new outlets or pop-up stores")
```
